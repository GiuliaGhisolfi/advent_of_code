### Part 1

```python
import csv

INPUT_DATA = "puzzle_input.csv"

with open(INPUT_DATA, newline='') as csvfile:
    reader = csv.reader(csvfile)
    data = list(reader)

    rules = data[:1176]
    ordered = data[1177:]
```

```python
order_list = []
for order in ordered:
    order_list.append([int(o) for o in order])
```

```python
rules_dict = {}
for rule in rules:
    first, second = rule[0].split("|")

    if first in rules_dict.keys():
        rules_dict[first].append(second)
    else:
        rules_dict[first] = [second]
```

```python
def follow_rules(order, rules_dict):
    for i, o in enumerate(order):
        if str(o) in rules_dict.keys():
            for s in rules_dict[str(o)]:
                if int(s) in order[:i]:
                    return 0
    
    #print(order[len(order)//2])
    return order[len(order)//2]

def main(order_list, rules_dict):
    total = 0
    for order in order_list:
        total += follow_rules(order, rules_dict)
    
    return total
```

```python
dummy_rules = {
    "97": ["13", "61", "47", "29", "53", "75"],
    "75": ["13", "29", "47", "53", "61"],
    "61": ["13", "29", "53"],
    "29": ["13", "29"],
    "53": ["13", "29"],
    "47": ["13", "29", "53", "61"],
}

dummy_order = [
    [75,47,61,53,29], # yes
    [97,61,53,29,13], # yes
    [75,29,13], # yes
    [75,97,47,61,53], # no
    [61,13,29], # no
    [97,13,75,29,47] # no
]

main(dummy_order, dummy_rules)
```

```python
main(order_list, rules_dict)
```

### Part 2

```python
def main_reordering(order_list, rules_dict):
    count = 0
    for order in order_list:
        if follow_rules(order, rules_dict) == 0:
            #print(order)
            count += reordering(order, rules_dict)

    return count

def reordering(order, rules_dict):
    reordered = order.copy()
    for i, o in enumerate(order):
        if str(o) in rules_dict.keys():
            for s in rules_dict[str(o)]:
                if int(s) in order[:i]:
                    reordered[i], reordered[reordered.index(int(s))] = reordered[reordered.index(int(s))], reordered[i]

    if follow_rules(reordered, rules_dict) == 0:
        return reordering(reordered, rules_dict)
    
    #print(f"Reordered: {reordered}")
    return reordered[len(reordered) // 2]
```

EXAMPLE:

75,97,47,61,53 becomes 97,75,47,61,53

61,13,29 becomes 61,29,13

97,13,75,29,47 becomes 97,75,47,29,13

```python
main_reordering(dummy_order, dummy_rules)
```

```python
main_reordering(order_list, rules_dict)
```
