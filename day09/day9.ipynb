### Part 1

```python
import csv

INPUT_DATA = "puzzle_input.csv"

with open(INPUT_DATA, newline='') as csvfile:
    reader = csv.reader(csvfile)
    data = list(reader)[0][0]

data
```

```python
def string_to_file_list(data):
    file = []
    idx = 0
    for i, val in enumerate(data):
        if i % 2 == 0:
            for _ in range(int(val)):
                file.append(idx)
            idx += 1
        else:
            for _ in range(int(val)):
                file.append(".")
    return file

file = string_to_file_list(data)
file
```

```python
def filesystem_checksum(file):
    checksum = 0
    for i, val in enumerate(file):
        if val == ".":
            break
        checksum += val * i
    return checksum

def compress_file(file):
    start = 0
    end = len(file) - 1

    while start < end:
        if file[start] == "." and file[end] != ".":
            file[start], file[end] = file[end], file[start]
            start += 1
            end -= 1
        else:
            if file[start] != ".":
                start += 1
            if file[end] == ".":
                end -= 1
    
    return file

def solution(data):
    file = string_to_file_list(data)
    file = compress_file(file)
    return filesystem_checksum(file)
```

```python
dummy_data = "2333133121414131402"

solution(dummy_data)
```

```python
solution(data)
```

```python
def new_filesystem_checksum(file):
    checksum = 0
    for i, val in enumerate(file):
        if val != ".":
            checksum += val * i
    return checksum

def new_compress_file(file):
    file_ids = sorted(set(x for x in file if x != "."), reverse=True)

    for file_id in file_ids:
        start = file.index(file_id)
        end = start
        while end + 1 < len(file) and file[end + 1] == file_id:
            end += 1

        file_size = end - start + 1

        free_start = -1
        free_size = 0
        for i in range(len(file)):
            if file[i] == ".":
                if free_start == -1:
                    free_start = i
                free_size += 1
                if free_size == file_size:
                    break
            else:
                free_start = -1
                free_size = 0

        if free_size == file_size and free_start < start:
            for i in range(file_size):
                file[free_start + i] = file_id
                file[start + i] = "."

    return file


def new_solution(data):
    file = string_to_file_list(data)
    print(file)
    file = new_compress_file(file)
    print(file)
    return new_filesystem_checksum(file)
```
