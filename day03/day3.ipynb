### Part 1

```python
import csv

INPUT_DATA = 'puzzle_input.csv'

with open(INPUT_DATA, newline='') as csvfile:
    data_csv = csv.reader(csvfile, delimiter='|')
    data_list = [row for row in data_csv]

len(data_list)
```

```python
data_list
```

```python
data = ''.join(''.join(i) for i in data_list)
data
```

```python
splitted_data = data.split("mul(")
len(splitted_data)
```

```python
splitted_data
```

```python
splitted_data = [elem.split(")")[0] for elem in splitted_data]

splitted_data
```

```python
res = 0

for elem in splitted_data:
    curr = elem.split(",")
    if len(curr) == 2:
        if len(curr[0]) <= 3 and len(curr[1]) <= 3:
            if str(int(curr[0])) == curr[0] and str(int(curr[1])) == curr[1]:
                res += int(curr[0]) * int(curr[1])

res
```

### Part 2

```python
donot_splitted_data = data.split("don't")
len(donot_splitted_data)
```

```python
new_data = donot_splitted_data[0]

for elem in donot_splitted_data[1:]:
    if "do" in elem:
        curr = elem.split("do")
        curr.pop(0)
        new_data += "do".join(curr)
```

```python
new_data
```

```python
new_splitted_data = new_data.split("mul(")
new_splitted_data = [elem.split(")")[0] for elem in new_splitted_data]

new_res = 0

for elem in new_splitted_data:
    curr = elem.split(",")
    if len(curr) == 2:
        if len(curr[0]) <= 3 and len(curr[1]) <= 3:
            if str(int(curr[0])) == curr[0] and str(int(curr[1])) == curr[1]:
                new_res += int(curr[0]) * int(curr[1])

new_res
```
