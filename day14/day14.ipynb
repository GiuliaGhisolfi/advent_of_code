```python
def final_position(x, y, vx, vy, height, width, delta_time):
    return (x + vx * delta_time) % width, (y + vy * delta_time) % height

def security_factor(pos, vel, height, width, delta_time=100):
    final_pos = [final_position(int(pos[i][0]), int(pos[i][1]), int(vel[i][0]), int(vel[i][1]),
        height, width, delta_time) for i in range(len(pos))]

    factor = 1

    factor *= sum(1 for x, y in final_pos if x < width // 2 and y < height // 2) # first quadrant
    factor *= sum(1 for x, y in final_pos if x >= (width // 2) + 1 and y < height // 2) # second quadrant
    factor *= sum(1 for x, y in final_pos if x < width // 2 and y >= (height // 2) + 1) # third quadrant
    factor *= sum(1 for x, y in final_pos if x >= (width // 2) + 1 and y >= (height // 2) + 1) # fourth quadrant

    return factor
```

```python
def find_tree_configuration(pos, vel, height, width, delta_time):
    final_pos = [final_position(int(pos[i][0]), int(pos[i][1]), int(vel[i][0]), int(vel[i][1]),
        height, width, delta_time) for i in range(len(pos))]
    
    final_pos.sort()

    prec = False
    count = 0
    for y in range(height):
        for x in range(width):
            if prec == False:
                if (x, y) in final_pos:
                    prec = True
            else:
                if (x, y) in final_pos:
                    count += 1
                else:
                    prec = False
                    count = 0
            if count == 10:
                return True
    return False

for t in range(10000):
    if find_tree_configuration(pos, vel, height, width, t):
        print(t)
        break
```

```python
def display_robot_conf(pos, vel, height, width, time):
    final_pos = [final_position(int(pos[i][0]), int(pos[i][1]), int(vel[i][0]), int(vel[i][1]),
        height, width, time) for i in range(len(pos))]

    matrix = [["." for x in range(width)] for y in range(height)]
    for y in range(height):
        for x in range(width):
            if (x, y) in final_pos:
                matrix[y][x] = "#"
                print("#", end="")
            else:
                print(".", end="")
        print()
```
