```python
from collections import deque

def manhattan_distance(x1, y1, x2, y2):
    return abs(x1 - x2) + abs(y1 - y2)

def find_min_distance(x_goal, y_goal, obstacles_coordinates, x_start=0, y_start=0):
    # Initialize the distance matrix
    distance_matrix = [[float('inf') for _ in range(x_goal+1)] for _ in range(y_goal+1)]
    distance_matrix[y_start][x_start] = 0

    # Mark obstacles
    for x, y in obstacles_coordinates:
        distance_matrix[y][x] = -1

    # Initialize the BFS queue
    queue = deque([(x_start, y_start)])

    # BFS to calculate shortest path
    while queue:
        x, y = queue.popleft()
        current_distance = distance_matrix[y][x]

        # Explore neighbors
        for dx, dy in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
            x_new, y_new = x + dx, y + dy

            # Check boundaries
            if x_new < 0 or x_new > x_goal or y_new < 0 or y_new > y_goal:
                continue

            # Skip obstacles or already visited cells
            if distance_matrix[y_new][x_new] == -1 or distance_matrix[y_new][x_new] <= current_distance + 1:
                continue

            # Update distance and enqueue
            distance_matrix[y_new][x_new] = current_distance + 1
            queue.append((x_new, y_new))

    # Return the shortest distance to the goal or -1 if unreachable
    final_distance = distance_matrix[y_goal][x_goal]
    return final_distance if final_distance != float('inf') else -1
```

```python
distance_matrix = [['.' for _ in range(x_max_dummy+1)] for _ in range(y_max_dummy+1)]
distance_matrix[0][0] = "."

# Mark obstacles
for x, y in dummy_coordinates[:12]:
    distance_matrix[y][x] = '#'

print("\n".join("".join(str(row)) for row in distance_matrix))
```

### Part 2

```python
def first_obastacle_to_block_path(x_goal, y_goal, obstacles_coordinates, x_start=0, y_start=0):
    for i, coords in enumerate(obstacles_coordinates):
        if find_min_distance(x_goal, y_goal, obstacles_coordinates[:i+1], x_start, y_start) == -1:
            return coords
    return None
```
